# **Projeto 03 — Relatório Comercial Interativo**

> Sobre o Projeto:
Este projeto foi desenvolvido durante meus estudos de Python e Análise de Dados.

> objetivo:
é transformar o Relatório Comercial Automatizado (Projeto 02) em uma aplicação interativa, permitindo que o usuário informe os dados da operação comercial para que o sistema calcule automaticamente os principais indicadores de desempenho (KPIs).

Todos os dados utilizados são fictícios e servem exclusivamente para fins de estudo e desenvolvimento de portfólio.

🎯 **Objetivo do Projeto**

Construir uma aplicação em Python capaz de receber dados comerciais informados pelo usuário e gerar automaticamente um relatório executivo com indicadores de desempenho.

Além da lógica de programação, o projeto busca aproximar o aprendizado de situações comuns encontradas na rotina de um Analista de Dados ou Analista Comercial.

📋 **Contexto**

Você foi contratado por uma empresa de consultoria comercial.

Todos os meses, o gerente da operação precisa calcular manualmente os indicadores de desempenho da empresa.

Sua missão é desenvolver uma aplicação em Python que automatize esse processo.

Ao executar o programa, o usuário deverá informar os dados do mês e, ao final, receber um relatório completo com todos os indicadores calculados automaticamente.

📥 **Dados que o usuário deverá informar**

O sistema solicitará as seguintes informações:

* Total de Agências Parceiras
* Agências que Realizaram Indicações
* Agências com Venda
* Total de Leads
* Total de Vendas
* Faturamento Total

📈 **Indicadores que serão calculados**
* Taxa de Ativação das Agências
* Percentual de Agências com Venda
* Taxa de Conversão
* Ticket Médio
* Percentual de Leads Não Convertidos
* Receita Média por Agência com Venda

```python
continuar = "S"

while continuar == "S":

    # ========================================
    # RELATÓRIO COMERCIAL AUTOMATIZADO
    # ========================================

    print("========================================")
    print("RELATÓRIO COMERCIAL AUTOMATIZADO")
    print("========================================")

    print()

    print("Bem-vindo!")
    print("Informe os dados abaixo para gerar o relatório comercial.")

    print()

    # Entrada de Dados

    total_agencias = int(input("Total de Agências Parceiras: "))
    agencias_com_indicacoes = int(input("Agências que realizaram indicações: "))
    agencias_com_vendas = int(input("Agências com venda: "))
    total_leads = int(input("Total de leads: "))
    total_vendas = int(input("Total de vendas realizadas: "))
    faturamento = float(input("Faturamento total (R$): "))

    # ==========================
    # Definição das Metas
    # ==========================

    meta_ativacao = 30
    meta_percentual = 20
    meta_conversao = 35
    meta_leads_nao_convertidos = 60

    # ==========================
    # Cálculo dos Indicadores
    # ==========================

    taxa_ativacao = (agencias_com_indicacoes / total_agencias) * 100
    percentual_agencias_venda = (agencias_com_vendas / total_agencias) * 100
    taxa_conversao = (total_vendas / total_leads) * 100
    ticket_medio = faturamento / total_vendas
    leads_nao_convertidos = total_leads - total_vendas
    percentual_leads_nao_convertidos = (leads_nao_convertidos / total_leads) * 100
    receita_media_por_agencia = faturamento / agencias_com_vendas

    # ==========================
    # RELATÓRIO EXECUTIVO
    # ==========================

    print()
    print("========================================")
    print("RELATÓRIO EXECUTIVO")
    print("========================================")

    print(f"Taxa de Ativação das Agências: {taxa_ativacao:.2f}%")

    if taxa_ativacao >= meta_ativacao:
        print("Meta de ativação das agências atingida.")
    else:
        print("Meta de ativação das agências não atingida.")

    print()

    print(f"Percentual de Agências com Venda: {percentual_agencias_venda:.2f}%")

    if percentual_agencias_venda >= meta_percentual:
        print("Meta de agências com venda atingida.")
    else:
        print("Meta de agências com venda não atingida.")

    print()

    print(f"Taxa de Conversão: {taxa_conversao:.2f}%")

    if taxa_conversao >= meta_conversao:
        print("Meta de conversão atingida.")
    else:
        print("Meta de conversão não atingida.")

    print()

    print(f"Ticket Médio: R$ {ticket_medio:.2f}")

    print()

    print(f"Percentual de Leads Não Convertidos: {percentual_leads_nao_convertidos:.2f}%")

    if percentual_leads_nao_convertidos >= meta_leads_nao_convertidos:
        print("Meta de leads não convertidos atingida.")
    else:
        print("Meta de leads não convertidos não atingida.")

    print()

    print(f"Receita Média por Agência com Venda: R$ {receita_media_por_agencia:.2f}")

    print()
    print("========================================")
    print("Relatório gerado com sucesso!")
    print("========================================")

    print()

    continuar = input("Deseja gerar um novo relatório? (S/N): ").upper()

print()
print("Programa encerrado. Até a próxima!")
```
